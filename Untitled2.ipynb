```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pickle
```

```python
# Creating a dictionary for sample data
data = {
    'Weather Condition': [1, 2, 3, 1, 2],
    'Road Type': [1, 2, 1, 2, 1],
    'Vehicle Speed': [80, 50, 60, 45, 70],
    'Traffic Density': [30, 45, 20, 50, 40],
    'Driver Experience': [5, 2, 10, 1, 3],
    'Accident Severity': [2, 3, 1, 2, 3]  # This is the dependent variable
}

# Creating the DataFrame
df = pd.DataFrame(data)
df
```

```python
# Independent variables
X = df[['Weather Condition', 'Road Type', 'Vehicle Speed', 'Traffic Density', 'Driver Experience']]
# Dependent variable
y = df['Accident Severity']

```

```python
# Initializing the model
model = LinearRegression()

# Fitting the model
model.fit(X, y)

# Displaying model coefficients
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
# Saving the model as a pickle file
with open('accident_severity_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Hypothetical data
hypothetical_data = np.array([[1, 2, 55, 35, 4]])

# Predicting accident severity
predicted_severity = model.predict(hypothetical_data)
print("Predicted Accident Severity:", predicted_severity[0])
```
